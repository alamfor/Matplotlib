```python
import pandas as pd
labels_medicine = ['Medicine(Before)','Medicine(After)']
labels_placebo = ['Placebo(Before)', 'Medicine(After)']
medicine = [2.88,1.42]
placebo = [2.72,2.68]
yerr_medicine = [0.12,0.08]
yerr_placebo = [0.24,0.13]

df = pd.DataFrame([medicine,placebo])
df.columns = ['Before','After']
df.index = ['Medicine','Placebo']

# Plot the bar chart with error bars
df.plot(kind='bar',width=0.4,color=['midnightblue','cornflowerblue'],yerr=[yerr_medicine,yerr_placebo])

plt.title('Effect of Medicines')
plt.xlabel('Condition')
plt.ylabel('[hsCRP] (mg/L)')
plt.xticks(rotation=0) # to keep the xtick labels horizontal
plt.legend(loc=(0.4,0.8))

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
framingham = pd.read_csv(r'framingham.csv')
```

```python
framingham.head()
```

```python
framingham = framingham.dropna()
```

```python
sns.countplot(framingham['education'])
```

```python
plt.figure(figsize=(7,5))
plt.title("Correlation among features sysBP , diaBP",fontsize=16)
sns.scatterplot(x='sysBP',y='diaBP',data=framingham)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.regplot(x='sysBP',y='diaBP',data=framingham)
plt.title("Correlation among features sysBP , diaBP",fontsize=16)
plt.show()
```

```python
sns.lmplot(x='sysBP',y='diaBP',hue='BPMeds',data=framingham)
plt.title("Correlation among features sysBP w.r.t BPMeds , diaBP",fontsize=16)
plt.show()
```
